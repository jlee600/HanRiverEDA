```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/ericjia/Desktop/School/CS 1302/Project/water_quality.csv")
df
```

```python
df.drop(columns = ['Site_Id', 'Unit_Id', 'Read_Date', 'Dissolved Oxygen (mg/L)', 'Secchi Depth (m)',
                   'Air Temp-Celsius', 'Time (24:00)', 'Field_Tech', 'DateVerified', 'WhoVerified',
                   'AirTemp (C)', 'Year'], inplace = True)
df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = new_df[:400]
new_df
```

```python
new_df[['Salinity (ppt)', 'pH (standard units)']].agg(['mean', 'median', 'var', 'std'])
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])

y1 = new_df['Water Temp (?C)']
y2 = new_df['Water Depth (m)']
y1_val = list(y1)
y2_val = list(y2)

x_val = list(range(1800,2200))

axe.plot(x_val, y1_val, color = 'r')
axe.plot(x_val, y2_val, color='b')
axe.set_xlabel('Year')
axe.set_ylabel('Water Stats')
axe.legend(labels=('Water Temp (C)','Water Depth (m)'))
axe.set_title('Water Temp VS. Water Depth')
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])

y1 = new_df['pH (standard units)']
y2 = new_df['Water Depth (m)']
y1_val = list(y1)
y2_val = list(y2)

x_val = list(range(1800,2200))

axe.plot(x_val, y1_val, color='grey')
axe.plot(x_val, y2_val, color = 'b')
axe.set_xlabel('Year')
axe.set_ylabel('Water Stats')
axe.legend(labels=('Salinity (ppt)','Water Depth (m)'))
axe.set_title('Salinity VS. Water Depth')
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])

y1 = new_df['Salinity (ppt)']
y2 = new_df['pH (standard units)']
y1_val = list(y1)
y2_val = list(y2)

x_val = list(range(1800,2200))

axe.scatter(x_val, y1_val, color='grey', s = 5)
axe.scatter(x_val, y2_val, color = 'y', s = 5)
axe.set_xlabel('Year')
axe.set_ylabel('Water Stats')
axe.legend(labels=('Salinity (ppt)','pH (standard units)'))
axe.set_title('Salinity VS. pH')
```

```python
def convert_fahrenheit_to_celsius(temp_list):
    celsius_list = []  
    
    for temp in temp_list:
        celsius = (temp - 32) * 5/9  
        celsius_list.append(celsius)  
    
    return celsius_list  

fig = plt.figure()
axe = fig.add_axes([0,0,1,1])

y1 = new_df['Water Temp (?C)']
y2 = new_df['Air Temp (?F)']
y1_val = list(y1)
y2_val_F = list(y2)
y2_val_C = convert_fahrenheit_to_celsius(y2_val_F)

x_val = list(range(1800,2200))

axe.scatter(x_val, y1_val, color='red', s = 5)
axe.scatter(x_val, y2_val_C, color = 'g', s = 5)
axe.set_xlabel('Year')
axe.set_ylabel('Water Stats')
axe.legend(labels=('Water Temp (C)','Air Temp (C)'))
axe.set_title('Water Temp (C) VS. Air Temp (C)')
```
